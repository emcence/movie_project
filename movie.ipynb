```python
#Calculate MDB's weighted rating (wr) which is given as :-
# WR = ((v/(v + m)) * R) + ((v/(v + m)) * C)

#where,

#v is the number of votes for the movie;
#m is the minimum votes required to be listed in the chart;
#R is the average rating of the movie; And
#C is the mean vote across the whole report
#We already have v(vote_count) and R (vote_average) and C can be calculated 

C= df2['vote_average'].mean()
print (C)

# mean rating for all the movies is approx 6 on a scale of 10

```

```python
m = df2['vote_count'].quantile(0.9)
print (m)
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
# Apply the weighted_rating function to create the 'score' feature
df2['score'] = df2.apply(lambda x: weighted_rating(x, m, C), axis=1)

# Print the updated DataFrame with the 'score' feature
print(df2.head())
```
